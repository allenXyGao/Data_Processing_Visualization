# 数据探索和样本构造

```python
# 针对label值做异常值处理往往很有效
```

```python
import numpy as np
import pandas as pd
from sklearn import preprocessing as prepc
from sklearn import feature_selection as fselect
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
X
```

```python
y = iris.target
y
```

# 离散特征 one Hot Encoding

```python
X_new = prepc.OneHotEncoder().fit_transform(y.reshape((-1, 1))).toarray()
X_new
```

# 特征标准化

```python
X_new = prepc.StandardScaler().fit_transform(X)
X_new
```

```python
X_minmax = prepc.MinMaxScaler().fit_transform(X)
X_minmax
```

# 数据变换

## 多项式

```python
prepc.PolynomialFeatures(degree=2).fit_transform(X)
```

```python
'''
多项式
X1, X2, X3, X4
获取幂次为0： 1
幂次为1: X1, X2, X3, X4，
幂次为2： X1^2, X1*X2, X1*X3,....
变成这么多加起来的特征
用处：FM模型
'''
```

## 对数log

#### log1p: log(1+x)

```python
'''
应用： 长尾分布
很少的值占据了很多的比例 eg 大部分人收入很低
“越大的值被压缩的越厉害”
'''
```

```python
prepc.FunctionTransformer(np.log1p, validate=True).fit_transform(X)
# validate 看是否合法 取log
```

## 二值化

```python
# 把一些连续的值变成离散的值，使得特征更好解释
```

```python
prepc.Binarizer(threshold=3).fit_transform(X)
```

# 特征筛选

```python
'''
1. 统计相关性
2. 基于正则L1
3. 基于树的重要性
4. 前向后向
'''
```

## 1. 统计相关性
方差： 标准化之前用，方差大的特征好
Chi-square 卡方检验
相关系数
互信息

```python
# 方差
fselect.VarianceThreshold(threshold=0.5).fit_transform(X)
```

```python
# 卡方 k=2, 保留两个
fselect.SelectKBest(fselect.chi2, k=2).fit_transform(X,y)
```

## 2.基于正则
对参数的惩罚
LR = Sigmoid(wx), 损失函数为log loss (or Sigmoid cross entropy)
-(ylog(fit)+(1-y)log(1-fit)) + Omega
###  L1 引起稀疏化， 系数容易变成0了， 说明这个对应的特征没有用了
why L1 正则容易稀疏化呢？

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(penalty="l1", C=0.1, multi_class='ovr', solver='liblinear').fit(X, y)
model = fselect.SelectFromModel(lr, prefit=True)
model.transform(X)
```

# 基于树的特征重要程度

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
fselect.SelectFromModel(GradientBoostingClassifier()).fit_transform(iris.data, iris.target)
```
